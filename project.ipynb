#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in Exercise 14) (★★★)

Print both values for each variable.

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(populations)]

mean_values = {
    'price': south_belt_subset['price'].mean(),
    'rooms': south_belt_subset['rooms'].mean(),
    'surface': south_belt_subset['surface'].mean(),
    'bathrooms': south_belt_subset['bathrooms'].mean()
}

variance_values = {
    'price': south_belt_subset['price'].var(),
    'rooms': south_belt_subset['rooms'].var(),
    'surface': south_belt_subset['surface'].var(),
    'bathrooms': south_belt_subset['bathrooms'].var()
}

print("Sample Mean Values:")
for variable, mean in mean_values.items():
    print(f"{variable}: {mean:.2f}")

print("\nSample Variance Values:")
for variable, variance in variance_values.items():
    print(f"{variable}: {variance:.2f}")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(populations)]

south_belt_subset['normalized_price'] = (south_belt_subset['price'] - south_belt_subset['price'].min()) / (south_belt_subset['price'].max() - south_belt_subset['price'].min())

plt.figure(figsize=(12, 8))

for population in populations:
    plt.hist(south_belt_subset[south_belt_subset['level5'] == population]['normalized_price'], 
             bins=20, 
             alpha=0.5, 
             label=population, 
             edgecolor='black')

plt.title('Normalized Price Distribution by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend(title='Population')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

The histograms highlight the normalized house price distributions for the residents of: Fuenlabrada, Leganés, Getafe, and Alcorcón. These populations are compared by depicting a standardised version of the price distributions making it possible to draw comparisons even when the original price distributions are different.

Observations:
Distribution Shape: Each population has its own spread of the normalized price of houses. For example, if one population’s maxima is shifted towards the right of the mean towards the higher normalized values, it implies that its median houses are, in fact, costlier than the median of more ‘populations’.

Density and Variability: The width of each histogram can be taken as the degree of variation of houses in the population being considered. Although rough, a more narrow histogram indicates a more uniform set of prices whilst a more dispersed one indicates a greater variation.

Comparative With Judgements: When the prices are normalized, we can make sense of the rank order of the populations. For instance, if there are more houses in Leganés with normalized prices in the upper range than in Fuenlabrada, it means Leganés is a ‘happening’ place.

Conclusion:
The purpose of this analysis is to gain a better understanding of the housing market dynamics within these populations. Knowing how normalized prices are distributed not only enables buyers to target but also assists buyers and investors in identifying geographical areas of interest due to comparative pricing trends and market states.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(populations)]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

for i, population in enumerate(populations):
    axes[i].scatter(south_belt_subset[south_belt_subset['level5'] == population]['surface'],
                    south_belt_subset[south_belt_subset['level5'] == population]['price'],
                    alpha=0.5)
    axes[i].set_title(f'Scatter Plot: {population}')
    axes[i].set_xlabel('Surface Area (m²)')
    axes[i].set_ylabel('Price (USD)')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(populations)]

coordinates_dict = {}

for population in populations:
    # Get unique latitude and longitude for each population
    latitudes = south_belt_subset[south_belt_subset['level5'] == population]['latitude'].unique()
    longitudes = south_belt_subset[south_belt_subset['level5'] == population]['longitude'].unique()
    
    if len(latitudes) > 0 and len(longitudes) > 0:
        coordinates_dict[population] = {
            'latitude': latitudes[0],
            'longitude': longitudes[0]
        }

coordinates_dict
```
